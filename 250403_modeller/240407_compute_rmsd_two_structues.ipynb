```python
# Install MDTraj
!pip install -q mdtraj

import mdtraj as md
import numpy as np

# Load the structures
ref = md.load_pdb('/content/4bgq.pdb')  # Now used as the reference
mobile = md.load_pdb('/content/01_cdkl5_af_1_303.pdb')  # Now aligned to 4bgq

# Select only CA atoms
ref_ca = ref.atom_slice(ref.top.select('name CA'))
mobile_ca = mobile.atom_slice(mobile.top.select('name CA'))

# Get residue indices
ref_res_ids = [atom.residue.index for atom in ref_ca.top.atoms]
mobile_res_ids = [atom.residue.index for atom in mobile_ca.top.atoms]

# Find common residues by index
common_res_ids = sorted(set(ref_res_ids).intersection(set(mobile_res_ids)))
print(f"Number of common residues (by index): {len(common_res_ids)}")

# Mask atom indices based on common residue indices
ref_mask = [i for i, atom in enumerate(ref_ca.top.atoms) if atom.residue.index in common_res_ids]
mobile_mask = [i for i, atom in enumerate(mobile_ca.top.atoms) if atom.residue.index in common_res_ids]

# Slice to common residues
ref_common = ref_ca.atom_slice(ref_mask)
mobile_common = mobile_ca.atom_slice(mobile_mask)

# Align reference structure to mobile (flip)
ref_common.superpose(mobile_common)

# Compute RMSD: now ref is aligned to mobile
rmsd = md.rmsd(ref_common, mobile_common)
print(f"\nRMSD (CA atoms, ref aligned to mobile): {rmsd[0]:.3f} Å")
```
