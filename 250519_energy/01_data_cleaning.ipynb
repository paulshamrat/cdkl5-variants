### 2.1 Comparison: Unique common old vs new clinvar

```python
import pandas as pd

# Load the two datasets
df_cdkl5_missense = pd.read_excel('00_data/clinvar_result_copy.xlsx', sheet_name='CDKL5_missense_AF')
df_cdkl5_condition = pd.read_excel('00_data/clinvar_result.xlsx', sheet_name='CDKL5_Condition')

# Extract 'Protein change' column from both datasets
protein_missense = set(df_cdkl5_missense['Protein change'].dropna())
protein_condition = set(df_cdkl5_condition['Protein change'].dropna())

# Find common and unique values
common_protein_changes = protein_missense.intersection(protein_condition)
unique_missense = protein_missense - protein_condition
unique_condition = protein_condition - protein_missense

# Convert results to DataFrames for saving
common_df = pd.DataFrame(list(common_protein_changes), columns=['Common Protein Changes'])
unique_missense_df = pd.DataFrame(list(unique_missense), columns=['Unique to CDKL5_missense_AF'])
unique_condition_df = pd.DataFrame(list(unique_condition), columns=['Unique to CDKL5_Condition'])

# Save comparisons in a new Excel file
comparison_file = "00_data/protein_change_comparison.xlsx"
with pd.ExcelWriter(comparison_file, mode='w', engine='openpyxl') as writer:
    common_df.to_excel(writer, sheet_name='Common_Protein_Changes', index=False)
    unique_missense_df.to_excel(writer, sheet_name='Unique_CDKL5_missense_AF', index=False)
    unique_condition_df.to_excel(writer, sheet_name='Unique_CDKL5_Condition', index=False)

print("Comparison results saved in 'protein_change_comparison.xlsx'.")
```

### 2.2 Comparison: Unique common old vs new clinvar after deleting 'intron variants'

```python
import pandas as pd

# Load the two datasets
df_cdkl5_missense = pd.read_excel('00_data/clinvar_result_copy.xlsx', sheet_name='CDKL5_missense_AF')
df_cdkl5_missense_only = pd.read_excel('00_data/clinvar_result.xlsx', sheet_name='CDKL5_missense_only')

# Extract 'Protein change' column from both datasets
protein_missense = set(df_cdkl5_missense['Protein change'].dropna())
protein_missense_only = set(df_cdkl5_missense_only['Protein change'].dropna())

# Find common and unique values
common_protein_changes = protein_missense.intersection(protein_missense_only)
unique_missense = protein_missense - protein_missense_only
unique_missense_only = protein_missense_only - protein_missense

# Create a summary dictionary
summary = {
    'Total in CDKL5_missense_AF': len(protein_missense),
    'Total in CDKL5_missense_only': len(protein_missense_only),
    'Total Common': len(common_protein_changes),
    'Total Unique to CDKL5_missense_AF': len(unique_missense),
    'Total Unique to CDKL5_missense_only': len(unique_missense_only)
}

# Print the summary
print("\nComparison Summary:")
for key, value in summary.items():
    print(f"{key}: {value}")

# Convert results to DataFrames for saving
summary_df = pd.DataFrame(list(summary.items()), columns=['Category', 'Count'])
common_df = pd.DataFrame(list(common_protein_changes), columns=['Common Protein Changes'])
unique_missense_df = pd.DataFrame(list(unique_missense), columns=['Unique to CDKL5_missense_AF'])
unique_missense_only_df = pd.DataFrame(list(unique_missense_only), columns=['Unique to CDKL5_missense_only'])

# Save comparisons in a new Excel file
comparison_file = "00_data/protein_change_comparison.xlsx"
with pd.ExcelWriter(comparison_file, mode='w', engine='openpyxl') as writer:
    summary_df.to_excel(writer, sheet_name='Comparison_Summary', index=False)
    common_df.to_excel(writer, sheet_name='Common_Protein_Changes', index=False)
    unique_missense_df.to_excel(writer, sheet_name='Unique_CDKL5_missense_AF', index=False)
    unique_missense_only_df.to_excel(writer, sheet_name='Unique_CDKL5_missense_only', index=False)

print("\nComparison results saved in 'protein_change_comparison.xlsx'.")
```

## 2.4 Protein_change_converted: p.Lys2Arg to K2R

```python
import pandas as pd
from openpyxl import load_workbook

# Full mapping of three-letter amino acid codes to one-letter codes
aa_mapping = {
    "Ala": "A", "Arg": "R", "Asn": "N", "Asp": "D",
    "Cys": "C", "Glu": "E", "Gln": "Q", "Gly": "G",
    "His": "H", "Ile": "I", "Leu": "L", "Lys": "K",
    "Met": "M", "Phe": "F", "Pro": "P", "Ser": "S",
    "Thr": "T", "Trp": "W", "Tyr": "Y", "Val": "V",
    "Sec": "U", "Pyl": "O", "Ter": "*"
}

# Load the Excel file and the specified sheet
file_path = '00_data/gnomAD.xlsx'
df_gnomad = pd.read_excel(file_path, sheet_name='Protein_Change_Converted')

# Strip 'p.' and store in a new column for parsing
df_gnomad['Protein Consequence (Three-Letter)'] = df_gnomad['Protein Consequence'].str.replace("p.", "", regex=False)

# Function to convert three-letter amino acid change to one-letter format
def convert_to_one_letter(protein_consequence):
    try:
        wt = protein_consequence[:3]                         # Wild-type (3-letter)
        mut = protein_consequence[-3:]                       # Mutant (3-letter)
        pos = protein_consequence[3:-3]                      # Position
        
        wt_1 = aa_mapping.get(wt, '?')
        mut_1 = aa_mapping.get(mut, '?')

        return f"{wt_1}{pos}{mut_1}"
    except Exception:
        return None

# Apply the function to create the corrected column
df_gnomad['Protein change (One-Letter)'] = df_gnomad['Protein Consequence (Three-Letter)'].apply(convert_to_one_letter)

# Save changes back to the same sheet (overwriting it)
with pd.ExcelWriter(file_path, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
    df_gnomad.to_excel(writer, sheet_name='Protein_Change_Converted', index=False)

print("Protein changes converted to one-letter format and saved in existing sheet 'Protein_Change_Converted'.")
```

### Sanity check one letter code

```python
import pandas as pd
import re

# Load the updated sheet
file_path = '00_data/gnomAD.xlsx'
df = pd.read_excel(file_path, sheet_name='Protein_Change_Converted')

# Define set of valid one-letter amino acid codes (including stop codon '*')
valid_aa = set('ACDEFGHIKLMNPQRSTVWY*')

# Sanity check function
def is_valid_change(change):
    if not isinstance(change, str):
        return False
    match = re.match(r'^([A-Z*])(\d+)([A-Z*])$', change)
    if match:
        wt, pos, mut = match.groups()
        return wt in valid_aa and mut in valid_aa
    return False

# Apply sanity check
df['Is_Valid_One_Letter_Change'] = df['Protein change (One-Letter)'].apply(is_valid_change)

# Report results
valid_count = df['Is_Valid_One_Letter_Change'].sum()
invalid_count = (~df['Is_Valid_One_Letter_Change']).sum()

print(f"\n✅ Valid entries: {valid_count}")
print(f"❌ Invalid entries: {invalid_count}")

# Show examples of invalid entries
if invalid_count > 0:
    print("\nExamples of invalid entries:")
    print(df.loc[~df['Is_Valid_One_Letter_Change'], ['Protein Consequence', 'Protein change (One-Letter)']].head())
```

## 5.9 missense_cdkl5_unique_prot_change_1L

```python
import pandas as pd

# File path and sheet names
file_path = '00_data/1kgp_cdkl5_grch38.xlsx'
input_sheet = 'missense_cdkl5_unique'
output_sheet = 'missense_cdkl5_unique_prot_chan'  # shortened sheet name, max 31 chars

# Load the Excel file and data
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name=input_sheet)

# Extract protein change info from 'HGVSp' column after ':p.'
df['Protein change 3L'] = df['HGVSp'].str.extract(r':p\.(.+)')

# Three-letter to one-letter AA code dictionary
aa_3to1 = {
    'Ala':'A', 'Arg':'R', 'Asn':'N', 'Asp':'D', 'Cys':'C',
    'Glu':'E', 'Gln':'Q', 'Gly':'G', 'His':'H', 'Ile':'I',
    'Leu':'L', 'Lys':'K', 'Met':'M', 'Phe':'F', 'Pro':'P',
    'Ser':'S', 'Thr':'T', 'Trp':'W', 'Tyr':'Y', 'Val':'V',
    'Sec':'U', 'Pyl':'O', 'Asx':'B', 'Glx':'Z', 'Xaa':'X'
}

# Function to convert 3-letter code protein change to 1-letter code
import re
def convert_3to1(prot_change_3L):
    if pd.isna(prot_change_3L):
        return prot_change_3L
    # Regex pattern to extract: wildtype (3 letters), position (number), mutant (3 letters)
    m = re.match(r'([A-Z][a-z]{2})(\d+)([A-Z][a-z]{2})', prot_change_3L)
    if not m:
        return prot_change_3L  # return as-is if pattern doesn't match
    wt_3, pos, mut_3 = m.groups()
    wt_1 = aa_3to1.get(wt_3, wt_3)
    mut_1 = aa_3to1.get(mut_3, mut_3)
    return f"{wt_1}{pos}{mut_1}"

# Apply conversion and add new column
df['Protein change'] = df['Protein change 3L'].apply(convert_3to1)

# Save to the same Excel file, replacing the output sheet if it exists
with pd.ExcelWriter(file_path, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
    df.to_excel(writer, sheet_name=output_sheet, index=False)

print(f"Saved updated dataframe with protein change columns to sheet '{output_sheet}'.")
```

## pull gnomAD allele frequency
data: clinvar
file: 00_data/clinvar_kgp_hector_unique_summary.xlsx
sheet: 'clinvar_1kgp_hector'
Protein change column: 'Protein change'
allele frequency column: 'Allele Frequency'

data: gnomAD
file: 00_data/gnomAD.xlsx
sheet: 'Protein_Change_Converted'
Protein change Column: 'Protein change (One-Letter)'

So what i want is:
read clinvar data, sheet and protein change column
then match with the entries found in 'Protein_Change_Converted' sheet of gnomAD data with the column 'Protien change (One-Letter)'

and pull out corresponding 'Allele Frequency'
and save it as a new sheet in the clinvar excel file with proper name


```python
import pandas as pd

# File paths
clinvar_file = '00_data/clinvar_kgp_hector_unique_summary.xlsx'
gnomad_file = '00_data/gnomAD.xlsx'

# Read ClinVar data
clinvar_df = pd.read_excel(clinvar_file, sheet_name='clinvar_1kgp_hector')
# Column with protein changes (one-letter code)
clinvar_col = 'Protein change'

# Read gnomAD data
gnomad_df = pd.read_excel(gnomad_file, sheet_name='Protein_Change_Converted')
gnomad_protein_col = 'Protein change (One-Letter)'
gnomad_af_col = 'Allele Frequency'

# Merge ClinVar with gnomAD on protein change (one-letter code)
merged_df = pd.merge(
    clinvar_df,
    gnomad_df[[gnomad_protein_col, gnomad_af_col]],
    left_on=clinvar_col,
    right_on=gnomad_protein_col,
    how='left'
)

# Rename the gnomAD Allele Frequency column for clarity
merged_df.rename(columns={gnomad_af_col: 'gnomAD Allele Frequency'}, inplace=True)

# Drop the redundant gnomAD protein change column after merge
merged_df.drop(columns=[gnomad_protein_col], inplace=True)

# Save to a new sheet in the same ClinVar Excel file
with pd.ExcelWriter(clinvar_file, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
    merged_df.to_excel(writer, sheet_name='clinvar_1kgp_hector_gaf', index=False)

print("Saved merged data with gnomAD Allele Frequency to sheet 'clinvar_1kgp_hector_gaf'.")
```

```python
import requests
import pandas as pd

# Step 1: Fetch CDKL5 sequence from UniProt (FASTA format)
uniprot_id = "O76039"
url = f"https://www.uniprot.org/uniprot/{uniprot_id}.fasta"
response = requests.get(url)
fasta_text = response.text

# Parse FASTA sequence (skip header line)
lines = fasta_text.split('\n')
sequence = ''.join(line.strip() for line in lines if not line.startswith('>'))

print(f"Length of CDKL5 sequence: {len(sequence)}")

# Step 2: Load your data with the 'Protein change' column
file_path = '00_data/clinvar_kgp_hector_unique_summary.xlsx'
sheet_name = 'clinvar_1kgp_hector_gaf'

df = pd.read_excel(file_path, sheet_name=sheet_name)

# Step 3: Extract wild type AA, position, mutant AA from 'Protein change'
def parse_protein_change(pc):
    # Example format: V107D
    if isinstance(pc, str) and len(pc) >= 3:
        wt = pc[0]
        mut = pc[-1]
        pos = pc[1:-1]
        if pos.isdigit():
            return wt, int(pos), mut
    return None, None, None

df[['wt_aa', 'pos', 'mut_aa']] = df['Protein change'].apply(
    lambda x: pd.Series(parse_protein_change(x))
)

# Step 4: Check match against UniProt sequence
def check_match(row):
    wt = row['wt_aa']
    pos = row['pos']
    if wt is None or pos is None or pos > len(sequence) or pos < 1:
        return 'Invalid'
    # UniProt sequence is 1-based index, python 0-based
    seq_aa = sequence[pos-1]
    return 'Match' if seq_aa == wt else f'Mismatch (Seq: {seq_aa})'

df['match_status'] = df.apply(check_match, axis=1)

# Step 5: Summary
summary = df['match_status'].value_counts()
print(summary)

# Optional: print mismatch rows to inspect
print("\nMismatches:")
print(df[df['match_status'].str.contains('Mismatch')][['Protein change', 'wt_aa', 'pos', 'mut_aa', 'match_status']])
```

```python
import pandas as pd

# File path and sheet names
file_path = '00_data/clinvar_kgp_hector_unique_summary.xlsx'
input_sheet = 'clinvar_1kgp_hector_gaf'
output_sheet = 'clinvar_1kgp_hector_gaf_final'

# Load the data
df = pd.read_excel(file_path, sheet_name=input_sheet)

# Remove mismatches based on exact 'Protein change'
mismatches = ['Y145H', 'A153G']
df_filtered = df[~df['Protein change'].isin(mismatches)].copy()

# Copy 'Protein change' column to new column 'Mutation'
df_filtered['Mutation'] = df_filtered['Protein change']

# Extract wild, position, and mutant from 'Mutation'
df_filtered['wild'] = df_filtered['Mutation'].str.extract(r'^([A-Z])')
df_filtered['position'] = df_filtered['Mutation'].str.extract(r'(\d+)').astype(int)
df_filtered['mutant'] = df_filtered['Mutation'].str.extract(r'([A-Z])$')

# Sort by position
df_filtered = df_filtered.sort_values(by='position')

# Save to a new sheet
with pd.ExcelWriter(file_path, mode='a', engine='openpyxl', if_sheet_exists='replace') as writer:
    df_filtered.to_excel(writer, sheet_name=output_sheet, index=False)

print(f"✅ Final dataset saved to sheet '{output_sheet}' in '{file_path}'.")
```
